sınıfta kaç kişi var gir sonra onların isimlerini ve notlarını al s
en düşük 2 . notu alanların isimlerini alfabetik olarak sırala




```python
data = []
score = []

for i in range(int(input())):           #burada sınıfta kaç kişi olduğu alınıyor sonra isimler ve notlar girdiriliyor ve sonra hepsi data ve score listesine atanıyor
  names = input()
  sc = int(input())
  score.append(sc)
  data.append([names,sc])


  '''
 burada notların atandığı liste tekrarlardan kurtulmak için önce set yapılıyor sonra
 büyükten küçüğe sorted ile sıralanıyor sonra second_low a en küçük ikinci atanıyor
 '''
s = sorted(list(set(score)))
second_low = s[1]

'''
burada boş bir liste oluşturuluyor sonra for döngüsünün data da kaç eleman varsa o kadar dönmesi sağlanıyor. sonra data listesindeki elementle birer liste olduğu için
her listenin ikinci index ine bak diyor ve bu second  low a eşitse o elemanın birinci indexi ni bu boş listeye eleman olarak ata diyor'''

selected = []
for i in range(len(data)):
  if data[i][1] == second_low:
    selected.append(data[i][0])

'''sonra bu selected listesini alfabetik olarak sıralıyor'''

selected.sort()

''' sonra bu listedeki elemanları istediği için for döngüsü ile bunları yazdırıyor'''
for i in selected:
  print(i,end='')






```

```python
n = int(input())
records = [[str(input()), float(input())] for i in range(n)]
second_lowest_score = sorted(set(i[1] for i in records))[1]
for i in sorted((i[0] for i in filter(lambda x: x[1] == second_lowest_score, records))):
  print(i)
```
